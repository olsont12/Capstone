## Scraping over wintering data adding 2 rows from current data from WWF

```python
overwintering = pd.read_html("https://www.fws.gov/midwest/monarch/OverwinteringMonarchs.html")
overwintering = overwintering[0]
overwintering.loc[25] = [2018, 14.9]
overwintering.loc[26] = [2019, 6.99]
```

```python
overwintering.columns = ['Year (data collected during December)', 'Total Area Covered (Acres)']
```

```python
overwintering['Population in Millons'] = round(overwintering['Total Area Covered (Acres)'] * 20.26, 2)
```

```python
overwintering = overwintering.melt(
    id_vars=['Year (data collected during December)'],
    var_name='Category',
    value_name='Metric'
)
```

```python
overwintering.to_csv("C:/Users/t_d_o/Documents/capstone/overwintering.csv")
```
